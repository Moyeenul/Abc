```python
import openpnm as op
import porespy as ps
import numpy as np
import matplotlib.pyplot as plt
op.visualization.set_mpl_style()
```

### Model Generation

```python
# Model Generation
a = [5, 8, 29]
pn = op.network.Cubic(shape = [5, 5], spacing=1e-6, connectivity=6)
op.topotools.trim(network=pn, throats=a)
```

### Visualization

```python
# Visualization
ax = op.visualization.plot_coordinates(pn)
ax = op.visualization.plot_connections(pn, ax = ax)
```

```python
# Declaring the pore and throat diameter
pn['pore.diameter'] = 2
pn['throat.diameter'] = 40print(pn)
```

```python
# Develop the 3D PTL
im = ps.generators.cylinders(shape = [200, 200, 200], porosity=0.6, r=3, theta_max=45)
```

```python
# Visualizing the 3D PTL

fig, ax = plt.subplots(1, 3, figsize=[18, 3])
ax[0].imshow(im[100, :, :]) # Image of x = 0
ax[1].imshow(im[:, 0, :]) # Image of y = 0
ax[2].imshow(im[:, :, 10]) # Image of z = 0
```
